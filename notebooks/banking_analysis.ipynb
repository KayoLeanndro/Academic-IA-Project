```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Realizando visualização iniciail dos dados

df = pd.read_csv('../data/raw/bank_marketing.csv', sep=',')
df.head()
```

```python
# verificando a existencia de campos vazios
df.isnull().sum()
```

```python
# Verificar dados duplicados
num_duplicated = df.duplicated().sum()

print(f"Número de linhas duplicadas: {num_duplicated}")
```

```python
# Verificar tipos de dados de cada coluna
data_types = df.dtypes
print("\nTipos de dados de cada coluna:")
print(data_types)
```

```python
# Salvar o DataFrame processado
processed_file_path = '../data/processed/bank_marketing_processed.csv'
df.to_csv(processed_file_path, index=False)
print(f"\nDataFrame processado salvo em: {processed_file_path}")
```

```python
sns.histplot(data=df, x='y')
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='job')
plt.xlabel('Profissão')
plt.ylabel('Contagem')
plt.xticks(rotation=45, ha='right')
plt.tight_layout() 
plt.show()
```
